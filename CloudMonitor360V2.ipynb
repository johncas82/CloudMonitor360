```python
#  Instalacin de paquetes necesarios para CloudMonitor360
!pip install pandas numpy matplotlib seaborn scikit-learn umap-learn pyspark \
pymongo confluent-kafka geopandas folium tensorflow flask flask-ngrok shap fairlearn
!pip install plotly
!pip install keras
!pip install scikeras
!pip install shap
!pip install shap fairlearn
!pip install shap fairlearn -q
#  Verificacin rpida de versiones instaladas
import pandas as pd
import numpy as np
import matplotlib
import seaborn
import sklearn
import umap
import pyspark
import pymongo
import tensorflow as tf
import flask
import shap
import fairlearn
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
from datetime import datetime, timedelta
import json
!pip install certifi
import certifi
from confluent_kafka import Producer, Consumer
import time
from collections import defaultdict
from pyspark.sql import SparkSession
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
import glob
from sklearn.cluster import KMeans
import plotly.express as px
from sklearn.model_selection import cross_val_score, StratifiedKFold, cross_validate
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, roc_auc_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from collections import Counter
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay
from tensorflow.keras.backend import clear_session
from fairlearn.metrics import (MetricFrame, demographic_parity_difference, demographic_parity_ratio)
from fairlearn.reductions import DemographicParity, ExponentiatedGradient
from sklearn.metrics import (accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, roc_curve, ConfusionMatrixDisplay)
import shap
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.metrics import roc_curve, ConfusionMatrixDisplay
from sklearn.model_selection import StratifiedKFold
shap.initjs()
from fairlearn.metrics import demographic_parity_difference, demographic_parity_ratio


print(" Todas las libreras han sido importadas correctamente.")
print(" pandas:", pd.__version__)
print(" numpy:", np.__version__)
print(" matplotlib:", matplotlib.__version__)
print(" seaborn:", seaborn.__version__)
print(" scikit-learn:", sklearn.__version__)
print(" umap-learn:", umap.__version__)
print(" PySpark:", pyspark.__version__)
print(" TensorFlow:", tf.__version__)
print(" Flask:", flask.__version__)
print(" SHAP:", shap.__version__)

#  Reproducibilidad
np.random.seed(42)

#  Parmetros de simulacin
n_registros = 1000
n_instancias = 10

#  Generacin de variables
df_ec2 = pd.DataFrame({
    "instance_id": np.random.choice(range(1, n_instancias + 1), size=n_registros),
    "timestamp": pd.date_range(start="2023-01-01", periods=n_registros, freq="H"),
    "cpu_utilization": np.random.normal(loc=50, scale=10, size=n_registros),
    "memory_used": np.random.normal(loc=60, scale=15, size=n_registros),
    "network_in": np.random.uniform(low=100, high=800, size=n_registros),
    "network_out": np.random.uniform(low=100, high=800, size=n_registros)
})

#  Guardar como CSV para pasos posteriores
df_ec2.to_csv("ec2_metrics.csv", index=False)

#  Verificacin rpida
print(" CSV generado con dimensiones:", df_ec2.shape)
df_ec2.head()


#  1. Cargar los datos
df = pd.read_csv("ec2_metrics.csv")

#  2. Eliminar filas con NaN
df.dropna(inplace=True)

#  2. Eliminar valores extremos por percentiles [1%, 99%]
metricas = ["cpu_utilization", "memory_used", "network_in", "network_out"]

for col in metricas:
    p1, p99 = np.percentile(df[col], [1, 99])
    df = df[(df[col] >= p1) & (df[col] <= p99)]

#  3. Validar la calidad del dataset
print(" Estructura del DataFrame:")
df.info()

print("\n Estadsticas descriptivas:")
print(df.describe())

print("\n Conteo de registros por instance_id:")
print(df["instance_id"].value_counts())

#  Cargar el CSV limpio
df = pd.read_csv("ec2_metrics.csv")

#  Asegurar tipo datetime para timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"])
df.sort_values("timestamp", inplace=True)
df.set_index("timestamp", inplace=True)

#  Histogramas de uso de CPU y memoria
plt.figure(figsize=(10,4))
sns.histplot(df["cpu_utilization"], kde=True, color="deepskyblue")
plt.title("Distribucin de CPU Utilization")
plt.xlabel("CPU (%)")
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x=df["memory_used"], color="lightgreen")
plt.title("Boxplot de Memoria Utilizada")
plt.xlabel("Memoria (%)")
plt.show()

#  Media mvil de 24h y deteccin de picos 3
df["cpu_ma"] = df["cpu_utilization"].rolling(window="24H").mean()
std_cpu = df["cpu_utilization"].std()
df["anomaly"] = df["cpu_utilization"] > (df["cpu_ma"] + 3 * std_cpu)

#  Serie temporal con picos resaltados
plt.figure(figsize=(14,6))
df["cpu_utilization"].plot(label="CPU Utilization", color="blue")
df["cpu_ma"].plot(label="Media mvil 24h", linestyle="--", color="orange")
if df["anomaly"].sum() > 0:
    df[df["anomaly"]]["cpu_utilization"].plot(style="o", color="red", label="Anomalas")
else:
    print("No se detectaron anomalas con el umbral actual.")
plt.title("Serie temporal de CPU con deteccin de picos")
plt.legend()
plt.xlabel("Timestamp")
plt.ylabel("CPU Utilization (%)")
plt.show()

#  Listado de timestamps anmalos
anomalous_timestamps = df[df["anomaly"]].index
print(" Timestamps con anomalas detectadas:")
print(anomalous_timestamps)

# Ruta al archivo de certificados raz confiables
ca_path = certifi.where()

#  URI de conexin con seguridad TLS explcita
uri = "mongodb+srv://johnalexcastiblanco:zIAGLjAZf2eSqtQC@cloudmonitor360.wujnyuv.mongodb.net/?retryWrites=true&w=majority&tls=true"

#  Conexin al clster
client = MongoClient(uri, tls=True, tlsCAFile=ca_path)

try:
    #  Verificar handshake y conexin
    server_info = client.server_info()
    print(" Conexin exitosa con MongoDB Atlas:")
    print(server_info)

except Exception as e:
    print(" Error de conexin:")
    print(e)
    #  Seleccin de base de datos y coleccin
    db = client["monitor"]
    col = db["ec2"]

    #  Insercin de DataFrame limpio
    df_reset = df.reset_index()
    registros = json.loads(df_reset.to_json(orient="records"))
    col.insert_many(registros)

    #  Crear ndice por instance_id
    col.create_index("instance_id")

    #  Consulta: instancias con CPU > 80% en ltimas 24 horas
    hace_24h = datetime.utcnow() - timedelta(hours=24)
    consulta = col.find({
        "cpu_utilization": { "$gt": 80 },
        "timestamp": { "$gte": hace_24h.isoformat() }
    })

    print(" Resultados: Instancias con CPU alta en ltimas 24h")
    for doc in consulta:
        print(doc)

except Exception as e:
    print(" Error al conectar o insertar en MongoDB Atlas:")
    print(e)


#  Parmetros de conexin al broker externo
conf_producer = {
    'bootstrap.servers': 'TU_BROKER_URL',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'TU_API_KEY',
    'sasl.password': 'TU_API_SECRET'
}

conf_consumer = {
    'bootstrap.servers': 'TU_BROKER_URL',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'TU_API_KEY',
    'sasl.password': 'TU_API_SECRET',
    'group.id': 'monitor_group',
    'auto.offset.reset': 'earliest'
}

#  1. Configurar Producer y emitir 200 mensajes JSON
producer = Producer(conf_producer)

df_sample = df.sample(200)  # muestra aleatoria
for _, row in df_sample.iterrows():
    msg = row.to_dict()
    producer.produce("ec2_metrics", value=json.dumps(msg))
producer.flush()
print(" Mensajes emitidos al tpico ec2_metrics")

#  2. Configurar Consumer y leer 10 segundos de mensajes
consumer = Consumer(conf_consumer)
consumer.subscribe(["ec2_metrics"])

cpu_agg = defaultdict(list)
start = time.time()
timeout = 10  # segundos

while time.time() - start < timeout:
    msg = consumer.poll(1.0)
    if msg is None:
        continue
    if msg.error():
        print(" Error en mensaje:", msg.error())
        continue
    try:
        data = json.loads(msg.value().decode("utf-8"))
        cpu_agg[data["instance_id"]].append(data["cpu_utilization"])
    except Exception as e:
        print(" Error procesando mensaje:", e)

consumer.close()

#  3. Calcular promedio de CPU por instancia
print(" Promedio de CPU por instancia (streaming en 10s):")
for inst, valores in cpu_agg.items():
    print(f" Instancia {inst}: {np.mean(valores):.2f}%")

#  Iniciar SparkSession en Google Colab
spark = SparkSession.builder \
    .appName("CloudMonitor360_ETL") \
    .getOrCreate()

#  Cargar el CSV simulado
df_spark = spark.read.csv("ec2_metrics.csv", header=True, inferSchema=True)

#  Verificacin rpida de schema
df_spark.printSchema()

#  Crear vista temporal para ejecutar SQL
df_spark.createOrReplaceTempView("metrics")

#  Ejecutar consulta agregada por instancia
query = """
    SELECT instance_id,
           ROUND(AVG(cpu_utilization),2) AS avg_cpu,
           ROUND(AVG(memory_used),2) AS avg_mem,
           ROUND(MAX(network_in),2) AS max_network_in,
           ROUND(MAX(network_out),2) AS max_network_out
    FROM metrics
    GROUP BY instance_id
"""

df_agg = spark.sql(query)
df_agg.coalesce(1).write.csv("ec2_agg", header=True, mode="overwrite")
df_agg.show()

#  Guardar resultado como CSV para clustering y visualizacin posterior
# df_agg.coalesce(1).write.csv("ec2_agg.csv", header=True, mode="overwrite")
# Buscar el archivo generado dentro de la carpeta ec2_agg
csv_path = glob.glob("/content/ec2_agg/part*.csv")[0]

# Cargarlo en pandas
df_agg = pd.read_csv(csv_path)

print(" CSV cargado con xito:", df_agg.shape)
df_agg.head()
print(" Archivo ec2_agg.csv exportado correctamente.")


# 1. Cargar ec2_agg.csv
#df_agg = pd.read_csv("/content/ec2_agg.csv")
print(" Dimensiones:", df_agg.shape)
df_agg.head()

#  UMAP en 3 dimensiones
X = df_agg.drop("instance_id", axis=1)
reductor_3d = umap.UMAP(n_components=3, random_state=42)
embeddings_3d = reductor_3d.fit_transform(X)

#  Clustering con KMeans
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)
labels = kmeans.labels_
score_kmeans = silhouette_score(X, labels)
print(f" Silhouette Score K-Means: {score_kmeans:.3f}")

#  Construir DataFrame reducido
df_reduced_3d = pd.DataFrame(embeddings_3d, columns=["x", "y", "z"])
df_reduced_3d["instance_id"] = df_agg["instance_id"]
df_reduced_3d["cluster"] = labels

#  (Opcional) Guardar como CSV
df_reduced_3d.to_csv("ec2_reduced_3d.csv", index=False)

#  Visualizacin interactiva
fig = px.scatter_3d(df_reduced_3d, x="x", y="y", z="z",
                    color="cluster", hover_data=["instance_id"],
                    title="Clusters EC2 en UMAP 3D")
fig.show()

# DBSCAN bsico
dbscan = DBSCAN(eps=30, min_samples=2).fit(X)
labels_dbscan = dbscan.labels_
# Filtrar ruido para calcular Silhouette slo si hay ms de 1 cluster
if len(set(labels_dbscan)) > 1 and -1 not in labels_dbscan:
    score_dbscan = silhouette_score(X, labels_dbscan)
    print(f" Silhouette Score DBSCAN: {score_dbscan:.3f}")
else:
    print(" DBSCAN no encontr suficientes clusters vlidos para calcular Silhouette.")

#  3. Reduccin de dimensin con UMAP para visualizacin
reductor = umap.UMAP(random_state=42)
embeddings = reductor.fit_transform(X)

# Construccin del DataFrame reducido
df_reduced = pd.DataFrame(embeddings, columns=["x", "y"])
df_reduced["instance_id"] = df_agg["instance_id"]
df_reduced["cluster"] = kmeans.labels_

#  4. Guardar resultado
#df_reduced.to_csv("ec2_reduced.csv", index=False)
print(" Archivo ec2_reduced.csv guardado.")

#  5. Scatter plot de clusters
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_reduced, x="x", y="y", hue="cluster", palette="tab10", s=100)
plt.title("Visualizacin 2D de Clusters por UMAP")
plt.xlabel("UMAP-X")
plt.ylabel("UMAP-Y")
plt.legend(title="Cluster")
plt.show()

#  Paso 7  Modelado predictivo (LogisticRegression + MLP)
#  Paso 7: Modelado predictivo  Validacin cruzada y mtricas
# Paso 7: Modelado Predictivo (20 min)

def paso7_modelado_predictivo(df_agg, rango_umbrales=(70, 20)):
    import numpy as np
    import matplotlib.pyplot as plt
    from collections import Counter
    from sklearn.preprocessing import StandardScaler
    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                            roc_auc_score, roc_curve, ConfusionMatrixDisplay)
    from sklearn.model_selection import StratifiedKFold
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense
    from tensorflow.keras.backend import clear_session

#  1. Copia del dataset y simulacin zona sensible
df_model = df_agg.copy()
df_model["zone"] = df_model["instance_id"].astype(int) % 2

#  2. Buscar umbral ptimo para generar al menos 2 clases
umbral_optimo = None
rango_umbrales = (70, 20)
for u in range(rango_umbrales[0], rango_umbrales[1]-1, -5):
    df_model["high_usage"] = (df_model["avg_cpu"] > u).astype(int)
    conteo = Counter(df_model["high_usage"])
    if len(conteo) >= 2:
        umbral_optimo = u
        print(f"\n Umbral ptimo encontrado: avg_cpu > {u}")
        print(" Distribucin de clases:", conteo)
        break

if umbral_optimo is None:
  print(" No se encontr un umbral que genere dos clases. Revisa los datos.")

#  3. Preparar X, y y X_scaled
y = df_model["high_usage"]
X = df_model.drop(["instance_id", "high_usage", "zone"], axis=1, errors="ignore")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#  4. Validacin cruzada con LogisticRegression
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores_lr = []

print("\n Evaluacin LogisticRegression por fold:")
for fold, (train_ix, val_ix) in enumerate(cv.split(X_scaled, y), 1):
    y_train = y.iloc[train_ix]
    if len(set(y_train)) < 2:
        print(f" Fold {fold}  Solo una clase en entrenamiento. Se omite.")
        continue

        model = LogisticRegression()
        model.fit(X_scaled[train_ix], y_train)
        y_pred = model.predict(X_scaled[val_ix])
        y_proba = model.predict_proba(X_scaled[val_ix])[:, 1]

        acc = accuracy_score(y.iloc[val_ix], y_pred)
        prec = precision_score(y.iloc[val_ix], y_pred)
        rec = recall_score(y.iloc[val_ix], y_pred)
        auc = roc_auc_score(y.iloc[val_ix], y_proba)

        scores_lr.append((acc, prec, rec, auc))
        print(f" Fold {fold}  Acc:{acc:.3f} Prec:{prec:.3f} Rec:{rec:.3f} AUC:{auc:.3f}")

    if scores_lr:
        scores_lr = np.array(scores_lr)
        print("\n Promedio LogisticRegression  Acc:", scores_lr[:,0].mean().round(3),
              "| Prec:", scores_lr[:,1].mean().round(3),
              "| Rec:", scores_lr[:,2].mean().round(3),
              "| AUC:", scores_lr[:,3].mean().round(3))

    #  5. Validacin cruzada con MLP
    def build_mlp(input_dim):
        model = Sequential()
        model.add(Dense(8, input_dim=input_dim, activation='relu'))
        model.add(Dense(4, activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        return model

    scores_mlp = []
    print("\n Evaluacin MLP por fold:")
    for fold, (train_ix, val_ix) in enumerate(cv.split(X_scaled, y), 1):
        y_train = y.iloc[train_ix]
        if len(set(y_train)) < 2:
            print(f" Fold {fold}  Solo una clase en entrenamiento. Se omite.")
            continue

        clear_session()
        model_mlp = build_mlp(X_scaled.shape[1])
        model_mlp.fit(X_scaled[train_ix], y_train, epochs=50, batch_size=8, verbose=0)
        y_pred = (model_mlp.predict(X_scaled[val_ix]) > 0.5).astype(int)
        y_proba = model_mlp.predict(X_scaled[val_ix]).flatten()

        acc = accuracy_score(y.iloc[val_ix], y_pred)
        prec = precision_score(y.iloc[val_ix], y_pred)
        rec = recall_score(y.iloc[val_ix], y_pred)
        auc = roc_auc_score(y.iloc[val_ix], y_proba)

        scores_mlp.append((acc, prec, rec, auc))
        print(f" Fold {fold}  Acc:{acc:.3f} Prec:{prec:.3f} Rec:{rec:.3f} AUC:{auc:.3f}")

    if scores_mlp:
        scores_mlp = np.array(scores_mlp)
        print("\n Promedio MLP  Acc:", scores_mlp[:,0].mean().round(3),
              "| Prec:", scores_mlp[:,1].mean().round(3),
              "| Rec:", scores_mlp[:,2].mean().round(3),
              "| AUC:", scores_mlp[:,3].mean().round(3))

    #  6. ROC y matriz de confusin con modelo final LR
    modelo_final = LogisticRegression()
    modelo_final.fit(X_scaled, y)
    y_pred_final = modelo_final.predict(X_scaled)
    y_proba_final = modelo_final.predict_proba(X_scaled)[:,1]

    fpr, tpr, _ = roc_curve(y, y_proba_final)
    auc_final = roc_auc_score(y, y_proba_final)

    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr, label=f"AUC = {auc_final:.2f}")
    plt.plot([0,1], [0,1], '--', color='gray')
    plt.title(" ROC Curve  LogisticRegression")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

    ConfusionMatrixDisplay.from_predictions(y, y_pred_final).plot()
    plt.title(" Matriz de Confusin  LogisticRegression")
    plt.show()


#Paso 8. Explicabilidad y Fairness
#  1. Calcular valores SHAP
# Definir nuevamente X y y para seguridad
#  Si no lo hiciste antes
!pip install shap fairlearn -q

#  Imports necesarios
import shap
from fairlearn.metrics import demographic_parity_difference, demographic_parity_ratio
from fairlearn.reductions import DemographicParity, ExponentiatedGradient

#  1. Clculo de valores SHAP
shap.initjs()

explainer = shap.Explainer(modelo_lr, X_scaled)
shap_values = explainer(X_scaled)

#  Grfico de barras SHAP
shap.plots.bar(shap_values, max_display=5)

#  2. Medicin de fairness antes de mitigacin
# Aseguramos que la columna sensible 'zone' est en df_model
sensitive = df_model["zone"]  # Simulacin de atributo sensible

#  Prediccin del modelo actual
y_pred = modelo_lr.predict(X_scaled)

#  Medicin de Demographic Parity antes de mitigacin
dp_diff = demographic_parity_difference(y_true=y, y_pred=y_pred, sensitive_features=sensitive)
dp_ratio = demographic_parity_ratio(y_true=y, y_pred=y_pred, sensitive_features=sensitive)

print(f"\n Fairness antes de mitigacin:")
print(f"- Dif:  {dp_diff:.3f}")
print(f"- Ratio:{dp_ratio:.3f}")

#  3. Mitigacin con Fairlearn: ExponentiatedGradient + DemographicParity
mitigador = ExponentiatedGradient(LogisticRegression(), constraints=DemographicParity())
mitigador.fit(X_scaled, y, sensitive_features=sensitive)

#  Nuevas predicciones mitigadas
y_mitigado = mitigador.predict(X_scaled)

#  Medicin post-mitigacin
dp_diff_post = demographic_parity_difference(y_true=y, y_pred=y_mitigado, sensitive_features=sensitive)
dp_ratio_post = demographic_parity_ratio(y_true=y, y_pred=y_mitigado, sensitive_features=sensitive)

print(f"\n Fairness despus de mitigacin:")
print(f"- Dif:  {dp_diff_post:.3f}")
print(f"- Ratio:{dp_ratio_post:.3f}")

#  Comparacin final
print("\n Comparativa de fairness:")
print(f"Antes  Dif: {dp_diff:.3f} | Ratio: {dp_ratio:.3f}")
print(f"Despus  Dif: {dp_diff_post:.3f} | Ratio: {dp_ratio_post:.3f}")

# Paso 9  Inspeccin y visualizacin de KPIs

#  1. Clculo de KPIs en df_agg
df_kpi = df_agg.copy()
df_kpi["high_usage"] = (df_kpi["avg_cpu"] > 70).astype(int)

#  Porcentaje total de alto uso
porcentaje_alto_uso = df_kpi["high_usage"].mean() * 100
print(f" Porcentaje de instancias con avg_cpu > 70%: {porcentaje_alto_uso:.2f}%")

#  Crear DataFrame de KPIs por instancia
ec2_kpis = df_kpi.groupby("instance_id")["high_usage"].mean().reset_index()
ec2_kpis["% high_usage"] = ec2_kpis["high_usage"] * 100
ec2_kpis.drop("high_usage", axis=1, inplace=True)

# 2. Bar plot de KPI por instancia

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
sns.barplot(data=ec2_kpis, x="instance_id", y="% high_usage", palette="Blues_r")
plt.xticks(rotation=45)
plt.title(" % de alto uso por instancia (avg_cpu > 70%)")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Instancia EC2")
plt.tight_layout()
plt.show()

#  3. Heatmap de correlaciones entre mtricas

#  Seleccionar mtricas agregadas
metricas = ["avg_cpu", "avg_mem", "max_network_in", "max_network_out"]
matriz_corr = df_agg[metricas].corr()

plt.figure(figsize=(6,5))
sns.heatmap(matriz_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title(" Correlacin entre mtricas de monitoreo EC2")
plt.show()

# Paso 10  Mdulo extendido de anlisis y visualizacin de KPIs CloudMonitor360

# Clculo de nuevos flags booleanos
df_kpi["high_mem"] = (df_kpi["avg_mem"] > 80).astype(int)
df_kpi["network_spike"] = ((df_kpi["max_network_in"] > 1_000_000) |
                           (df_kpi["max_network_out"] > 1_000_000)).astype(int)

# Consolidacin de KPIs por instancia
kpi_cols = ["high_usage", "high_mem", "network_spike"]
ec2_kpis_ext = df_kpi.groupby("instance_id")[kpi_cols].mean().reset_index()
ec2_kpis_ext[[f"%_{col}" for col in kpi_cols]] = ec2_kpis_ext[kpi_cols] * 100
ec2_kpis_ext.drop(columns=kpi_cols, inplace=True)

# Vista tabular
print(ec2_kpis_ext.head())

# 2. Visualizacin de KPIs (exportable en .png)
#a. KPIs comparativos por instancia
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12,4))
sns.barplot(data=ec2_kpis_ext, x="instance_id", y="%_high_usage", palette="viridis", ax=ax)
plt.xticks(rotation=45)
plt.title(" % CPU alto por instancia EC2")
plt.ylabel("Porcentaje (%)")
plt.xlabel("Instancia")
plt.tight_layout()
plt.savefig("kpi_high_cpu.png")  #  Exportable
plt.show()

#b. KPIs comparativos por instancia
fig, axs = plt.subplots(1, 3, figsize=(16, 4))
for i, col in enumerate(["%_high_usage", "%_high_mem", "%_network_spike"]):
    sns.barplot(data=ec2_kpis_ext, x="instance_id", y=col, ax=axs[i], palette="mako")
    axs[i].set_title(f"{col} por instancia")
    axs[i].tick_params(axis='x', rotation=45)
    axs[i].set_ylabel("%")
plt.tight_layout()
plt.savefig("kpi_comparativos.png")  #  Exportable
plt.show()

#c.  3. Heatmap extendido de correlaciones
# Mtricas originales + flags
metricas_ext = ["avg_cpu", "avg_mem", "max_network_in", "max_network_out",
                "high_usage", "high_mem", "network_spike"]
corr_ext = df_kpi[metricas_ext].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_ext, annot=True, cmap="Spectral", fmt=".2f")
plt.title(" Heatmap extendido de correlaciones")
plt.savefig("heatmap_correlaciones.png")  #  Exportable
plt.show()

# Estructura sugerida para blog tcnico / notebook reproducible

# CloudMonitor360  Monitoreo tico y visualizacin reproducible

## Introduccin
#Breve presentacin del reto tcnico, arquitectura cloud y valores ticos (explicabilidad, impacto).

## Integracin cloud y ETL
#Descripcin del pipeline con MongoDB Atlas, PySpark, transformacin y agregacin.

## KPI Insights y visualizacin
#Explicacin de KPIs clave, thresholds de alerta, visualizaciones exportables (.png) con insights accionables.

## Fairness y explicabilidad
# Mencin de posibles sesgos en uso de recursos, importancia de visualizaciones
# claras para decisin tica.

## Conclusiones y prximos pasos
#Impacto del monitoreo tico, mejoras futuras (alertas automatizadas, dashboards interactivos, blog FUNDAPUCS).

###  Paso 11 - Sincronizacin automtica con GitHub (desde Colab)
#Permite enviar los archivos del paquete `CloudMonitor360_package` en Drive al repositorio `CloudMonitor360` en GitHub va comandos Git.
#Ideal para mantener control de versiones reproducible sin usar interfaces grficas externas.

---
```
